# 4. Data Preprocessing and Feature Engineering {#data-preprocessing}
# 4.1 Data Loading and Initial Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import warnings
warnings.filterwarnings('ignore')

# Load datasets (already loaded in your notebook)
temp_datasets = {
    '-10C': low_curr_ocv_minus_10,
    '0C': low_curr_ocv_0,
    '10C': low_curr_ocv_10,
    '20C': low_curr_ocv_20,
    '25C': low_curr_ocv_25,
    '30C': low_curr_ocv_30,
    '40C': low_curr_ocv_40,
    '50C': low_curr_ocv_50
}

# Combine all temperature datasets with temperature labels
combined_data = []
for temp, dataset in temp_datasets.items():
    dataset_copy = dataset.copy()
    dataset_copy['Temperature_Label'] = temp
    dataset_copy['Temperature_Numeric'] = float(temp.replace('C', ''))
    combined_data.append(dataset_copy)

# Create master dataset
master_df = pd.concat(combined_data, ignore_index=True)
print(f"Master dataset shape: {master_df.shape}")
print(f"Temperature distribution:\n{master_df['Temperature_Label'].value_counts()}")
```

# 4.2 Missing Data Analysis

```python
# Check for missing values
missing_data = master_df.isnull().sum()
missing_percentage = (missing_data / len(master_df)) * 100

print("Missing Data Analysis:")
print("="*50)
for col, missing_count in missing_data.items():
    if missing_count > 0:
        print(f"{col}: {missing_count} ({missing_percentage[col]:.2f}%)")

# Handle missing values using time-series appropriate methods
# Forward fill for slowly changing variables
master_df['Voltage(V)'].fillna(method='ffill', inplace=True)
master_df['Current(A)'].fillna(method='ffill', inplace=True)

# Interpolate for continuous variables
master_df['Temperature (C)_1'].interpolate(method='linear', inplace=True)
if 'Temperature (C)_2' in master_df.columns:
    master_df['Temperature (C)_2'].interpolate(method='linear', inplace=True)
```

# 4.3 Advanced Feature Engineering

```python
def create_battery_features(df):
    """
    Create advanced battery features following the methodology from 
    'Advanced Feature Engineering for A123 Battery Analysis'
    """
    features_df = df.copy()
    
    # 1. Derived Power and Energy Features
    features_df['Instantaneous_Power'] = features_df['Voltage(V)'] * features_df['Current(A)']
    features_df['Energy_Efficiency'] = (features_df['Discharge_Energy(Wh)'] / 
                                       (features_df['Charge_Energy(Wh)'] + 1e-8))
    features_df['Coulombic_Efficiency'] = (features_df['Discharge_Capacity(Ah)'] / 
                                          (features_df['Charge_Capacity(Ah)'] + 1e-8))
    
    # 2. Temperature-Dependent Features
    features_df['Voltage_Temp_Coeff'] = features_df['Voltage(V)'] / (features_df['Temperature_Numeric'] + 273.15)
    features_df['Resistance_Temp_Factor'] = features_df['Internal_Resistance(Ohm)'] * np.exp(1000 / (features_df['Temperature_Numeric'] + 273.15))
    
    # 3. Statistical Window Features (5-point rolling windows)
    for window in [5, 20]:
        features_df[f'Voltage_Mean_{window}'] = features_df['Voltage(V)'].rolling(window=window).mean()
        features_df[f'Voltage_Std_{window}'] = features_df['Voltage(V)'].rolling(window=window).std()
        features_df[f'Current_Mean_{window}'] = features_df['Current(A)'].rolling(window=window).mean()
        features_df[f'Power_Mean_{window}'] = features_df['Instantaneous_Power'].rolling(window=window).mean()
    
    # 4. Polynomial Features for Non-linear Behavior
    from sklearn.preprocessing import PolynomialFeatures
    poly_features = ['Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)']
    X_poly = features_df[poly_features].values
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_poly_transformed = poly.fit_transform(X_poly)
    
    # Add polynomial features
    poly_feature_names = poly.get_feature_names_out(poly_features)
    for i, name in enumerate(poly_feature_names):
        if name not in poly_features:  # Avoid duplicating original features
            features_df[f'Poly_{name}'] = X_poly_transformed[:, i]
    
    # 5. Temporal Features
    features_df['Time_Since_Start'] = features_df['Test_Time(s)'] - features_df['Test_Time(s)'].iloc[0]
    features_df['Step_Duration'] = features_df['Step_Time(s)']
    
    # 6. State of Charge Approximation
    features_df['SOC_Approx'] = features_df['Charge_Capacity(Ah)'] / features_df['Charge_Capacity(Ah)'].max()
    
    # 7. Impedance and Phase Features
    features_df['Impedance_Real'] = features_df['AC_Impedance(Ohm)'] * np.cos(np.radians(features_df['ACI_Phase_Angle(Deg)']))
    features_df['Impedance_Imag'] = features_df['AC_Impedance(Ohm)'] * np.sin(np.radians(features_df['ACI_Phase_Angle(Deg)']))
    
    return features_df

# Apply feature engineering
print("Applying advanced feature engineering...")
engineered_df = create_battery_features(master_df)

# Remove any remaining NaN values
engineered_df = engineered_df.dropna()
print(f"Engineered dataset shape: {engineered_df.shape}")
```

# 4.4 Battery Health Classification Target Creation

```python
def create_health_labels(df):
    """
    Create battery health labels based on capacity and efficiency metrics
    Following industry standards for battery health assessment
    """
    # Calculate health metrics
    max_discharge_capacity = df['Discharge_Capacity(Ah)'].max()
    df['Capacity_Retention'] = df['Discharge_Capacity(Ah)'] / max_discharge_capacity
    
    # Health classification based on capacity retention and efficiency
    conditions = [
        (df['Capacity_Retention'] >= 0.90) & (df['Energy_Efficiency'] >= 0.85),
        (df['Capacity_Retention'] >= 0.80) & (df['Capacity_Retention'] < 0.90),
        df['Capacity_Retention'] < 0.80
    ]
    
    choices = ['Healthy', 'Moderate', 'Critical']
    df['Battery_Health'] = np.select(conditions, choices, default='Moderate')
    
    return df

# Create health labels
engineered_df = create_health_labels(engineered_df)

# Display health distribution
health_distribution = engineered_df['Battery_Health'].value_counts()
print("Battery Health Distribution:")
print("="*40)
print(health_distribution)
print(f"Health percentages:\n{engineered_df['Battery_Health'].value_counts(normalize=True) * 100}")
```

# 4.5 Feature Selection and Scaling

```python
# Select relevant features for modeling
feature_columns = [
    'Voltage(V)', 'Current(A)', 'Internal_Resistance(Ohm)',
    'Charge_Capacity(Ah)', 'Discharge_Capacity(Ah)',
    'Charge_Energy(Wh)', 'Discharge_Energy(Wh)',
    'dV/dt(V/s)', 'AC_Impedance(Ohm)', 'ACI_Phase_Angle(Deg)',
    'Temperature_Numeric', 'Instantaneous_Power', 'Energy_Efficiency',
    'Coulombic_Efficiency', 'Voltage_Temp_Coeff', 'Resistance_Temp_Factor',
    'Voltage_Mean_5', 'Voltage_Std_5', 'Current_Mean_5', 'Power_Mean_5',
    'SOC_Approx', 'Impedance_Real', 'Impedance_Imag'
]

# Add available polynomial features
poly_cols = [col for col in engineered_df.columns if col.startswith('Poly_')]
feature_columns.extend(poly_cols)

# Remove any features that don't exist in the dataframe
available_features = [col for col in feature_columns if col in engineered_df.columns]
print(f"Using {len(available_features)} features for modeling")

# Prepare data for modeling
X = engineered_df[available_features].copy()
y = engineered_df['Battery_Health'].copy()

# Handle any remaining missing values
X = X.fillna(X.mean())

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"Feature matrix shape: {X.shape}")
print(f"Target distribution: {np.bincount(y_encoded)}")
```

# 5. Classification Model Implementation {#classification-models}
# 5.1 Data Splitting with Temperature Stratification

```python
# Avoid data leakage by ensuring temperature balance in train/test splits
from sklearn.model_selection import StratifiedShuffleSplit

# Create stratified split considering both health status and temperature
stratify_feature = engineered_df['Temperature_Label'] + '_' + engineered_df['Battery_Health']

# Split data while maintaining temperature and health distribution
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
train_idx, test_idx = next(sss.split(X, stratify_feature))

X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y_encoded[train_idx], y_encoded[test_idx]

print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")

# Verify temperature distribution in splits
train_temps = engineered_df.iloc[train_idx]['Temperature_Label'].value_counts()
test_temps = engineered_df.iloc[test_idx]['Temperature_Label'].value_counts()
print(f"Training temperature distribution:\n{train_temps}")
print(f"Test temperature distribution:\n{test_temps}")
```

# 5.2 Feature Scaling

```python
# Apply different scaling techniques as mentioned in requirements
scalers = {
    'StandardScaler': StandardScaler(),
    'MinMaxScaler': MinMaxScaler()
}

scaled_data = {}
for scaler_name, scaler in scalers.items():
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    scaled_data[scaler_name] = {
        'X_train': X_train_scaled,
        'X_test': X_test_scaled,
        'scaler': scaler
    }
    print(f"{scaler_name} applied successfully")
```

# 5.3 Model 1: Support Vector Machine (SVM)

```python
import time
from sklearn.model_selection import learning_curve

def implement_svm_classifier():
    """
    Implement SVM classifier with proven default parameters
    Following the SVM principles from 'In Depth: Support Vector Machines'
    Uses battery-analytics optimized defaults for immediate deployment
    """
    print("Implementing SVM Classifier with Default Parameters...")
    print("="*50)
    
    # Use StandardScaler for SVM (performs better with standardized features)
    X_train_svm = scaled_data['StandardScaler']['X_train']
    X_test_svm = scaled_data['StandardScaler']['X_test']
    
    # SVM with proven defaults for battery health classification
    svm_model = SVC(
        C=1.0,                    # Good default for most battery datasets
        gamma='scale',            # Automatically adapts to feature count (29 features)
        kernel='rbf',             # Best for non-linear battery electrochemistry
        probability=True,         # Enable probability estimates for health confidence
        class_weight='balanced',  # Handle potential class imbalance
        random_state=42
    )
    
    print("SVM Configuration:")
    print(f"  C: {svm_model.C} (regularization parameter)")
    print(f"  Gamma: {svm_model.gamma} (RBF kernel coefficient)")
    print(f"  Kernel: {svm_model.kernel} (for non-linear battery behavior)")
    print(f"  Class Weight: {svm_model.class_weight} (balanced classes)")
    
    # Train the model
    start_time = time.time()
    svm_model.fit(X_train_svm, y_train)
    training_time = time.time() - start_time
    
    # Predictions
    y_pred_svm = svm_model.predict(X_test_svm)
    y_pred_proba_svm = svm_model.predict_proba(X_test_svm)
    
    # Performance metrics
    accuracy_svm = accuracy_score(y_test, y_pred_svm)
    f1_svm = f1_score(y_test, y_pred_svm, average='weighted')
    
    print(f"\nSVM Results:")
    print(f"  Training Time: {training_time:.1f} seconds")
    print(f"  Support Vectors: {svm_model.n_support_}")
    print(f"  Total Support Vectors: {sum(svm_model.n_support_)}")
    print(f"  Support Vector Ratio: {sum(svm_model.n_support_)/len(X_train_svm)*100:.1f}%")
    print(f"  Accuracy: {accuracy_svm:.4f}")
    print(f"  F1-Score: {f1_svm:.4f}")
    
    # Generate SVM visualizations
    print(f"\nGenerating SVM training visualizations...")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Learning Curve (SVM Training Progress)
    train_sizes, train_scores, val_scores = learning_curve(
        svm_model, X_train_svm, y_train, 
        train_sizes=np.linspace(0.1, 1.0, 8),
        cv=3, n_jobs=-1, random_state=42
    )
    
    train_mean = train_scores.mean(axis=1)
    train_std = train_scores.std(axis=1)
    val_mean = val_scores.mean(axis=1)
    val_std = val_scores.std(axis=1)
    
    ax1.plot(train_sizes, train_mean, 'b-', linewidth=2, label='Training Score', marker='o')
    ax1.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2, color='blue')
    
    ax1.plot(train_sizes, val_mean, 'r-', linewidth=2, label='Validation Score', marker='s')
    ax1.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.2, color='red')
    
    ax1.set_xlabel('Training Set Size')
    ax1.set_ylabel('Accuracy Score')
    ax1.set_title('SVM Learning Curve\n(Training Progress)', fontweight='bold')
    ax1.legend(loc='lower right')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0.7, 1.0)
    
    # Add convergence annotation
    final_gap = abs(train_mean[-1] - val_mean[-1])
    ax1.annotate(f'Final Gap: {final_gap:.3f}\nConverged: {"Yes" if final_gap < 0.05 else "No"}',
                xy=(train_sizes[-1], val_mean[-1]),
                xytext=(train_sizes[-2], val_mean[-1] - 0.05),
                arrowprops=dict(arrowstyle='->', color='green'),
                fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))
    
    # Plot 2: Decision Function Distribution
    decision_scores = svm_model.decision_function(X_test_svm)
    
    # For multi-class SVM, decision_function returns a matrix
    if len(decision_scores.shape) > 1:
        # Use the maximum decision score for each sample
        decision_scores = np.max(decision_scores, axis=1)
    
    ax2.hist(decision_scores, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
    ax2.axvline(x=0, color='red', linestyle='--', linewidth=2, label='Decision Boundary')
    ax2.set_xlabel('Decision Function Score')
    ax2.set_ylabel('Frequency')
    ax2.set_title('SVM Decision Function Distribution', fontweight='bold')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Add statistics
    ax2.text(0.02, 0.98, f'Mean: {decision_scores.mean():.3f}\nStd: {decision_scores.std():.3f}\nRange: [{decision_scores.min():.2f}, {decision_scores.max():.2f}]',
             transform=ax2.transAxes, verticalalignment='top',
             bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Support Vector Analysis
    print(f"\nSupport Vector Analysis:")
    class_names = ['Critical', 'Healthy', 'Moderate']  # Assuming these are the classes
    for i, (class_name, n_sv) in enumerate(zip(class_names, svm_model.n_support_)):
        percentage = (n_sv / sum(svm_model.n_support_)) * 100
        print(f"  {class_name}: {n_sv} support vectors ({percentage:.1f}%)")
    
    return svm_model, y_pred_svm, y_pred_proba_svm, accuracy_svm, f1_svm

# Run SVM implementation
svm_model, svm_predictions, svm_probabilities, svm_accuracy, svm_f1 = implement_svm_classifier()
```

# 5.4 Model 2: Random Forest

```python
from sklearn.model_selection import learning_curve

def implement_random_forest():
    """
    Implement Random Forest classifier with proven default parameters
    Based on 'In Depth: Decision Trees and Random Forests' principles
    Optimized defaults for battery health classification across temperatures
    """
    print("\nImplementing Random Forest Classifier with Default Parameters...")
    print("="*50)
    
    # Use MinMaxScaler for Random Forest (handles mixed feature scales well)
    X_train_rf = scaled_data['MinMaxScaler']['X_train']
    X_test_rf = scaled_data['MinMaxScaler']['X_test']
    
    # Random Forest with battery-analytics optimized defaults
    rf_model = RandomForestClassifier(
        n_estimators=200,         # Sufficient trees for 174K samples
        max_depth=20,             # Prevent overfitting while capturing complexity
        min_samples_split=5,      # Conservative splitting for stability
        min_samples_leaf=2,       # Smooth decision boundaries
        max_features='sqrt',      # Standard for classification (29  5 features)
        bootstrap=True,           # Enable bagging for variance reduction
        oob_score=True,           # Enable out-of-bag scoring for progress tracking
        class_weight='balanced',  # Handle potential class imbalance
        random_state=42,
        n_jobs=-1                 # Use all available CPU cores
    )
    
    print("Random Forest Configuration:")
    print(f"  Trees: {rf_model.n_estimators}")
    print(f"  Max Depth: {rf_model.max_depth} (prevents overfitting)")
    print(f"  Min Samples Split: {rf_model.min_samples_split}")
    print(f"  Max Features: {rf_model.max_features} (~5 features per split)")
    print(f"  OOB Score: {rf_model.oob_score} (training progress tracking)")
    print(f"  Class Weight: {rf_model.class_weight}")
    print(f"  Parallel Jobs: {rf_model.n_jobs} (uses all CPU cores)")
    
    # Train the model
    start_time = time.time()
    rf_model.fit(X_train_rf, y_train)
    training_time = time.time() - start_time
    
    # Predictions
    y_pred_rf = rf_model.predict(X_test_rf)
    y_pred_proba_rf = rf_model.predict_proba(X_test_rf)
    
    # Performance metrics
    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    f1_rf = f1_score(y_test, y_pred_rf, average='weighted')
    
    # Feature importance analysis (bonus with Random Forest)
    feature_importance = pd.DataFrame({
        'feature': available_features,
        'importance': rf_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print(f"\nRandom Forest Results:")
    print(f"  Training Time: {training_time:.1f} seconds")
    print(f"  Out-of-Bag Score: {rf_model.oob_score_:.4f}")
    print(f"  Accuracy: {accuracy_rf:.4f}")
    print(f"  F1-Score: {f1_rf:.4f}")
    
    # Plot 1: OOB Error Progression (Random Forest Training Progress)
    print(f"\nGenerating Random Forest training visualizations...")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # OOB error progression
    oob_errors = []
    tree_range = range(10, 201, 10)  # Every 10 trees to speed up
    
    for n_trees in tree_range:
        rf_temp = RandomForestClassifier(
            n_estimators=n_trees, 
            oob_score=True, 
            max_depth=20,
            min_samples_split=5,
            random_state=42,
            n_jobs=-1
        )
        rf_temp.fit(X_train_rf, y_train)
        oob_errors.append(1 - rf_temp.oob_score_)
    
    ax1.plot(tree_range, oob_errors, 'b-', linewidth=2, marker='o', markersize=4)
    ax1.set_xlabel('Number of Trees')
    ax1.set_ylabel('Out-of-Bag Error Rate')
    ax1.set_title('Random Forest Training Progress\n(OOB Error vs Trees)', fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0, max(oob_errors) * 1.1)
    
    # Add annotation for optimal point
    min_error_idx = np.argmin(oob_errors)
    min_error_trees = tree_range[min_error_idx]
    min_error_value = oob_errors[min_error_idx]
    ax1.annotate(f'Optimal: {min_error_trees} trees\nOOB Error: {min_error_value:.3f}',
                xy=(min_error_trees, min_error_value),
                xytext=(min_error_trees + 30, min_error_value + 0.01),
                arrowprops=dict(arrowstyle='->', color='red'),
                fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))
    
    # Plot 2: Feature Importance
    top_features = feature_importance.head(10)
    ax2.barh(range(len(top_features)), top_features['importance'], color='lightblue')
    ax2.set_yticks(range(len(top_features)))
    ax2.set_yticklabels(top_features['feature'], fontsize=9)
    ax2.set_xlabel('Feature Importance')
    ax2.set_title('Top 10 Feature Importances', fontweight='bold')
    ax2.grid(axis='x', alpha=0.3)
    
    # Add values on bars
    for i, v in enumerate(top_features['importance']):
        ax2.text(v + 0.001, i, f'{v:.3f}', va='center', fontsize=9)
    
    plt.tight_layout()
    plt.show()
    
    print(f"Top 10 Most Important Features:")
    for i, row in feature_importance.head(10).iterrows():
        print(f"  {row['feature']:25s}: {row['importance']:.4f}")
    
    return rf_model, y_pred_rf, y_pred_proba_rf, accuracy_rf, f1_rf, feature_importance

# Run Random Forest implementation
rf_model, rf_predictions, rf_probabilities, rf_accuracy, rf_f1, rf_feature_importance = implement_random_forest()
```

```python

def implement_neural_network():
    """
    Implement Neural Network classifier with proven default parameters
    Following the ANN methodology from 'Implementation of the ANN'
    Optimized architecture for battery health classification
    """
    print("\nImplementing Neural Network Classifier with Default Parameters...")
    print("="*50)
    
    # Use StandardScaler for ANN (neural networks prefer normalized inputs)
    X_train_ann = scaled_data['StandardScaler']['X_train']
    X_test_ann = scaled_data['StandardScaler']['X_test']
    
    # Neural Network with battery-analytics optimized defaults
    ann_model = MLPClassifier(
        hidden_layer_sizes=(100, 50),  # Two hidden layers: input(25)  100  50  output(3)
        activation='relu',             # ReLU works well with battery physics features
        solver='adam',                 # Adaptive learning rate optimizer
        alpha=0.001,                   # Light regularization to prevent overfitting
        learning_rate='adaptive',      # Reduce learning rate when loss improvement < tol
        learning_rate_init=0.001,      # Initial learning rate
        max_iter=1000,                 # Maximum iterations (usually converges < 500)
        early_stopping=True,           # Stop when validation score stops improving
        validation_fraction=0.1,       # 10% of training data for validation
        n_iter_no_change=10,           # Patience: stop if no improvement for 10 iterations
        tol=1e-4,                      # Tolerance for optimization
        random_state=42
    )
    
    print("Neural Network Architecture:")
    print(f"  Input Layer: {len(available_features)} features")
    print(f"  Hidden Layer 1: {ann_model.hidden_layer_sizes[0]} neurons (ReLU)")
    print(f"  Hidden Layer 2: {ann_model.hidden_layer_sizes[1]} neurons (ReLU)")
    print(f"  Output Layer: 3 neurons (Healthy/Moderate/Critical)")
    print(f"  Solver: {ann_model.solver} (adaptive learning rate)")
    print(f"  Regularization (): {ann_model.alpha}")
    print(f"  Early Stopping: {ann_model.early_stopping}")
    
    # Train the model
    start_time = time.time()
    ann_model.fit(X_train_ann, y_train)
    training_time = time.time() - start_time
    
    # Predictions
    y_pred_ann = ann_model.predict(X_test_ann)
    y_pred_proba_ann = ann_model.predict_proba(X_test_ann)
    
    # Performance metrics
    accuracy_ann = accuracy_score(y_test, y_pred_ann)
    f1_ann = f1_score(y_test, y_pred_ann, average='weighted')
    
    print(f"\nNeural Network Results:")
    print(f"  Training Time: {training_time:.1f} seconds")
    print(f"  Converged in: {ann_model.n_iter_} iterations")
    print(f"  Final Loss: {ann_model.loss_:.6f}")
    print(f"  Accuracy: {accuracy_ann:.4f}")
    print(f"  F1-Score: {f1_ann:.4f}")
    
    # Plot learning curve if available
    if hasattr(ann_model, 'loss_curve_') and len(ann_model.loss_curve_) > 1:
        plt.figure(figsize=(10, 6))
        plt.plot(ann_model.loss_curve_, 'b-', linewidth=2, label='Training Loss')
        plt.title('Neural Network Training Progress', fontsize=14, fontweight='bold')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.grid(True, alpha=0.3)
        plt.legend()
        
        # Add convergence info
        plt.text(0.7, 0.9, f'Converged in {ann_model.n_iter_} epochs', 
                transform=plt.gca().transAxes, fontsize=12,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.5))
        plt.show()
        
        print(f"  Training converged successfully (early stopping activated)")
    
    return ann_model, y_pred_ann, y_pred_proba_ann, accuracy_ann, f1_ann

# Run Neural Network implementation
ann_model, ann_predictions, ann_probabilities, ann_accuracy, ann_f1 = implement_neural_network()
```

# 7. Cross-Temperature Validation {#cross-temperature-validation}
# 7.1 Advanced Validation Strategy

```python
def cross_temperature_validation():
    """
    Implement cross-temperature validation to test model generalization
    Following the validation methodology from 'Hyperparameters and Model Validation'
    Uses default parameters for efficient cross-temperature testing
    """
    print("Cross-Temperature Validation Analysis:")
    print("="*50)
    
    temperatures = list(temp_datasets.keys())
    validation_results = []
    
    for hold_out_temp in temperatures:
        print(f"\nTesting generalization - Holding out {hold_out_temp}")
        
        # Create training set excluding hold-out temperature
        train_temps = [t for t in temperatures if t != hold_out_temp]
        
        # Prepare training data (all temperatures except hold-out)
        train_data = []
        for temp in train_temps:
            temp_data = engineered_df[engineered_df['Temperature_Label'] == temp]
            train_data.append(temp_data)
        
        train_combined = pd.concat(train_data, ignore_index=True)
        
        # Prepare test data (only hold-out temperature)
        test_data = engineered_df[engineered_df['Temperature_Label'] == hold_out_temp]
        
        if len(test_data) > 100:  # Ensure sufficient test samples
            # Features and targets
            X_train_cv = train_combined[available_features].fillna(train_combined[available_features].mean())
            y_train_cv = label_encoder.transform(train_combined['Battery_Health'])
            
            X_test_cv = test_data[available_features].fillna(test_data[available_features].mean())
            y_test_cv = label_encoder.transform(test_data['Battery_Health'])
            
            # Scale features
            scaler_cv = StandardScaler()
            X_train_cv_scaled = scaler_cv.fit_transform(X_train_cv)
            X_test_cv_scaled = scaler_cv.transform(X_test_cv)
            
            # Use Random Forest with default parameters (fastest and most reliable)
            rf_cv = RandomForestClassifier(
                n_estimators=200,
                max_depth=20,
                min_samples_split=5,
                min_samples_leaf=2,
                max_features='sqrt',
                bootstrap=True,
                class_weight='balanced',
                random_state=42,
                n_jobs=-1
            )
            
            # Train and predict
            start_time = time.time()
            rf_cv.fit(X_train_cv_scaled, y_train_cv)
            training_time = time.time() - start_time
            
            y_pred_cv = rf_cv.predict(X_test_cv_scaled)
            
            # Calculate metrics
            accuracy_cv = accuracy_score(y_test_cv, y_pred_cv)
            f1_cv = f1_score(y_test_cv, y_pred_cv, average='weighted')
            
            validation_results.append({
                'Hold_Out_Temperature': hold_out_temp,
                'Train_Temperatures': ', '.join(train_temps),
                'Test_Samples': len(test_data),
                'Training_Time_Sec': training_time,
                'Accuracy': accuracy_cv,
                'F1-Score': f1_cv
            })
            
            print(f"  Training Time: {training_time:.1f}s")
            print(f"  Accuracy: {accuracy_cv:.4f}, F1-Score: {f1_cv:.4f}")
    
    # Create results dataframe
    cv_results_df = pd.DataFrame(validation_results)
    
    print(f"\nCross-Temperature Validation Summary:")
    print("="*60)
    print(cv_results_df[['Hold_Out_Temperature', 'Test_Samples', 'Training_Time_Sec', 'Accuracy', 'F1-Score']])
    
    print(f"\nMean Cross-Temperature Performance:")
    print(f"Average Training Time: {cv_results_df['Training_Time_Sec'].mean():.1f}  {cv_results_df['Training_Time_Sec'].std():.1f} seconds")
    print(f"Average Accuracy: {cv_results_df['Accuracy'].mean():.4f}  {cv_results_df['Accuracy'].std():.4f}")
    print(f"Average F1-Score: {cv_results_df['F1-Score'].mean():.4f}  {cv_results_df['F1-Score'].std():.4f}")
    print(f"Total Cross-Temperature Validation Time: {cv_results_df['Training_Time_Sec'].sum():.1f} seconds")
    
    return cv_results_df

# Run cross-temperature validation
cross_temp_results = cross_temperature_validation()
```

# 8.2 Model Robustness Analysis

```python
def analyze_model_robustness():
    """Analyze model robustness and limitations"""
    
    print("Model Robustness Analysis:")
    print("="*40)
    
    # Calculate prediction confidence for each model
    def analyze_prediction_confidence(probabilities, predictions, model_name):
        max_probs = np.max(probabilities, axis=1)
        mean_confidence = np.mean(max_probs)
        
        # Identify low confidence predictions
        low_confidence_mask = max_probs < 0.7
        low_confidence_count = np.sum(low_confidence_mask)
        
        print(f"\n{model_name} Confidence Analysis:")
        print(f"  Average prediction confidence: {mean_confidence:.4f}")
        print(f"  Low confidence predictions (<0.7): {low_confidence_count} ({low_confidence_count/len(predictions)*100:.1f}%)")
        
        return mean_confidence, low_confidence_count
    
    # Analyze each model
    svm_conf, svm_low = analyze_prediction_confidence(svm_probabilities, svm_predictions, "SVM")
    rf_conf, rf_low = analyze_prediction_confidence(rf_probabilities, rf_predictions, "Random Forest")
    ann_conf, ann_low = analyze_prediction_confidence(ann_probabilities, ann_predictions, "Neural Network")
    
    # Model agreement analysis
    agreement_mask = (svm_predictions == rf_predictions) & (rf_predictions == ann_predictions)
    agreement_rate = np.mean(agreement_mask)
    
    print(f"\nModel Agreement Analysis:")
    print(f"  All models agree: {np.sum(agreement_mask)} predictions ({agreement_rate*100:.1f}%)")
    print(f"  Disagreement cases: {np.sum(~agreement_mask)} predictions ({(1-agreement_rate)*100:.1f}%)")
    
    # Analyze disagreement cases
    disagreement_indices = np.where(~agreement_mask)[0]
    if len(disagreement_indices) > 0:
        print(f"\nDisagreement Analysis:")
        print(f"  Cases where models disagree often indicate:")
        print(f"  - Boundary conditions between health states")
        print(f"  - Measurement noise or outliers")
        print(f"  - Need for additional features")
    
    return {
        'confidence_scores': [svm_conf, rf_conf, ann_conf],
        'agreement_rate': agreement_rate,
        'disagreement_count': len(disagreement_indices)
    }

# Analyze model robustness
robustness_results = analyze_model_robustness()
```
